```python
#Syed Mahmood
#Python Pyodide Kernel JupyterLite
#AdventOfCode2023 Day Two Solution

#Part One


#Open the file with Python
my_file = open("input2.txt", "r")
#Read the file into a string
data = my_file.read()
#Convert the string into a list object by splitting for each newline
data_converted_to_list = data.split("\n")
#Remove the element (empty string)
data_converted_to_list.pop()
data_converted_to_list
```

```python
#Track the upper bound for each color
highest_colors = {
    "red" : 12,
    "blue" : 14,
    "green" : 13
    }

sum_of_games = 0

#Loop through each game
for game in data_converted_to_list:
    #Obtain the Number of the game by calling two string splits
    game_number = int(game.split(":")[0].split(" ")[1])
    #Obtain a list of each run in the game
    run_list = game.split(':')[1].split(";")
    #Set a boolean for determining if we need to add the game number to our final sum of games
    add=True
    #Iterate through each round and check if they exceed our bounds we set
    for run in run_list:
        #Check each color individually
        if "," in run:
            info = run.split(",")
            for color in info:
                #Remove the first char which is always a space
                color_cleaned = color[1:]
                #Get a small two element list of the color and count
                color_info = color_cleaned.split(" ")
                #Check the count
                if int(color_info[0]) > highest_colors[color_info[1]]:
                    #Set our boolean to False if the count exceeds our upper bound
                    add = False
        
        #To account for only one color in the round            
        else:
            #Remove the first char which is always a space
            color_cleaned = run[1:]
            #Get a small two element list of the color and count
            color_info = color_cleaned.split(" ")
            #Check the count
            if int(color_info[0]) > highest_colors[color_info[1]]:
                #Set our boolean to False if the count exceeds our upper bound
                add = False
    
    #Add our game number if it didn't exceed the bounds
    if add:
        sum_of_games = sum_of_games + game_number
    
print(sum_of_games)
```

```python
#Part Two

sum_of_games = 0

#Loop through each game
for game in data_converted_to_list:
    #Track the highest number of colors found in each game
    highest_colors = {
        "red" : 0,
        "blue" : 0,
        "green" : 0
        }
    #Obtain the Number of the game by calling two string splits
    game_number = int(game.split(":")[0].split(" ")[1])
    #Obtain a list of each run in the game
    run_list = game.split(':')[1].split(";")
    #Iterate through each round and replace the highest counts if they exceed the previous counts
    for run in run_list:
        #Check each color individually
        if "," in run:
            info = run.split(",")
            for color in info:
                #Remove the first char which is always a space
                color_cleaned = color[1:]
                #Get a small two element list of the color and count
                color_info = color_cleaned.split(" ")
                #Check the count against our previous highest count
                if int(color_info[0]) > highest_colors[color_info[1]]:
                    #Set the new highest value in our dictionary
                    highest_colors[color_info[1]] = int(color_info[0])
        
        #To account for only one color in the round           
        else:
            #Remove the first char which is always a space
            color_cleaned = run[1:]
            #Get a small two element list of the color and count
            color_info = color_cleaned.split(" ")
            #Check the count against our previous highest count
            if int(color_info[0]) > highest_colors[color_info[1]]:
                #Set the new highest value in our dictionary
                highest_colors[color_info[1]] = int(color_info[0])
    
    #Multiply the highest values in our dictionary
    power = 1
    for num in list(highest_colors.values()):
        power = power * num
    #Add the value to our sum of games
    sum_of_games = sum_of_games + power
    
print(sum_of_games)
```
